```
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
import random
```

```
def gausian_function(x,variance,mean):
  mu=mean
  sigma=math.sqrt(variance)
  y=(1/(sigma*math.sqrt(2*np.pi)))*((np.exp([1])[0])**(-0.5*(((x-mu)/sigma)**2)))
  return y
```

```
gausian_function(5,1,5)
```

# This shows all the likelihood functions with changing means

```
x1_points=[]
y1_points=[]
list_of_points=[]
variance=1
mean=5
for x in range(11):
  y=gausian_function(x,variance,mean)
  x1_points.append(x)
  y1_points.append(y)
  list_of_points.append([x,y])
plt.scatter(x1_points,y1_points)
plt.title("Distribution of 10 sample points ")  
plt.figure(figsize = (20,10))
plt.subplots_adjust(bottom=2, top=3, hspace=1)

for i in range(0,11,1):
  x_points=[]
  y_points=[]
  list_of_points=[]
  variance=1
  mean=i
  for x in range(11):
    y=gausian_function(x,variance,mean)
    x_points.append(x)
    y_points.append(y)
    list_of_points.append([x,y])
  plt.subplot(3,4,i+1)  
  plt.scatter(x1_points,y1_points)  
  plt.plot(x_points,y_points)
  plt.title('For mean = '+ str(mean) + ", Variance = " + str(variance))
  plt.xlabel('x')
  plt.ylabel('y')
```

## Likelihood Function values plotted v/s Means (fitted with varying means)

```
likelihood_function_values_for_each_mean=[]
for mean in range(0,11,1):
  variance=1
  function_value=1  # Value just initialized
  for x in np.arange(0,10,0.5):
    y=gausian_function(x,variance,mean)
    if(x==1):
      function_value=y
    else:
      function_value=function_value*y
  likelihood_function_values_for_each_mean.append(function_value)
```

```
X=[x for x in range(11)]
Y=likelihood_function_values_for_each_mean
```
